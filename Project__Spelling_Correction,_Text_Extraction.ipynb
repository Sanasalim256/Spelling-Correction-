# **Spelling Correction**

### **Import the dependency**

```python
from textblob import TextBlob
```

```python
text = 'i am goood'
textb = TextBlob(text)
textb.correct().string
```

## **Another way to do this**

```python
from spellchecker import SpellChecker
```

```python
spell = SpellChecker()

text = 'i aam goood'

dict_of_autocorrect_words = {}
for i in spell.unknown(text.split()):
    dict_of_autocorrect_words[i] = spell.correction(i)

print(f'The AUTOCORRECT suggestions are Mis-spelled words are {dict_of_autocorrect_words}')

temp = text.split()
res = []
for wrd in temp:

    res.append(dict_of_autocorrect_words.get(wrd, wrd))

res = ' '.join(res)

print(res)
```

```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8, 16
```

```python
reader = easyocr.Reader(['en'])
img = '/content/apj.jpg'
Image(img)
```

```python
output = reader.readtext(img)
output
```

```python
# Text correction
text = ''
for i in range(len(output)):
  text =text +' '+ output[i][1]

print(text)
```
